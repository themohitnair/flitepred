```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/enriched_dataset.csv")
df.drop(["CANCELLED"], inplace=True, axis="columns")
df.head()
```

```python
feature_cols = ["MONTH", "DAY_OF_MONTH", "HOUR", "TIME_OF_DAY", "DAY_OF_WEEK", "IS_WEEKEND", "MKT_CARRIER", "DEP_TIME", "DISTANCE", "TEMPERATURE", "HUMIDITY", "WIND_SPEED", "PRECIPITATION"]
X = df[feature_cols]
Y = df["DEP_DELAY"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

X.hist(figsize=(14, 10), bins=30, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[feature_cols + ['DEP_DELAY']].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(data=df, x="TIME_OF_DAY", y="DEP_DELAY")
plt.title("Departure Delay by Time of Day")
plt.xlabel("Time of Day (1=Morning, ..., 4=Night)")
plt.ylabel("Departure Delay (minutes)")
plt.show()
```

```python
sns.boxplot(data=df, x="IS_WEEKEND", y="DEP_DELAY")
plt.title("Departure Delay: Weekday vs Weekend")
plt.xlabel("Is Weekend")
plt.ylabel("Departure Delay (minutes)")
plt.show()
```

```python
sns.scatterplot(x='HUMIDITY', y='DEP_DELAY', data=df, alpha=0.3)
plt.title("Humidity vs Departure Delay")
plt.show()

sns.scatterplot(x='WIND_SPEED', y='DEP_DELAY', data=df, alpha=0.3)
plt.title("Wind Speed vs Departure Delay")
plt.show()
```
